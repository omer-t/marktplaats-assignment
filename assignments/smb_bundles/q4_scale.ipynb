# Question 4: Marketing Campaign Proposal

## Assignment
> The introduction of SMB Bundles has provided value to both Marktplaats and our SMB sellers, as they are now more visible on our platform and are able to distinguish themselves from others.
>
> Without access to additional data, we would like you to think conceptually and answer the following:
>
> Imagine you are tasked with promoting the new SMB bundles more broadly. Propose one marketing campaign, including the target audience, the channel(s) you would use, and the key KPI(s) you would focus on to measure success.


## 1. Campaign Objective

- Promote bundles to active sellers with a high SMB likelihood score (from question #1).
- Start with sellers who also have a high readiness score (from question #2), then expand to broader seller groups.

Potential future campaigns:
- Reactivation campaign for SMB sellers with declining performance.
- Upsell campaign from Basic to Plus.


## 2. Target Audience

Base audience: active sellers without a bundle.

Use the SMB-likelihood score and readiness-score to set message intensity:

| Segment | Audience | Campaign role |
|---|---|---|
| Tier 1 | High SMB likelihood + high readiness | Direct conversion push: upgrade now |
| Tier 2 | High SMB likelihood + medium readiness | Education and nurture: explain bundle benefits |
| Tier 3 | Medium SMB likelihood + medium/low readiness | Light-touch awareness and interest validation |
| Excluded | Low SMB likelihood | Avoid intrusive promotion to likely consumers |

Start with Tier 1 for the strongest adoption impact, then broaden to Tier 2 and Tier 3 through scalable CRM and in-platform prompts.


## 3. Channels

### 3.1 Lifecycle CRM

Use email / CRM journeys triggered by seller behavior.

Examples of triggers:

| Trigger | Purpose |
|---|---|
| Seller lists repeatedly | Promote bundle when commercial intent is high |
| Seller uses paid features | Position bundle as a structured visibility product |
| Seller operates in competitive category | Promote differentiation benefit |
| Seller receives buyer interest but has no bundle | Promote visibility/professionalization as next step |

Apply by tier:

| Tier | When | Frequency |
|---|---|---|
| 1 | Any strong trigger | Full journey, capped over first 2 weeks |
| 2 | Repeated or high-value trigger | Limited nurture, lower frequency |
| 3 | Test cell only | One touch |

### 3.2 In-platform Prompts

Use prompts in high-intent seller surfaces:

| Surface | Why this moment works |
|---|---|
| Seller dashboard | Natural moment to link bundles to better performance |
| Listing creation flow | Bundle can improve listings from day one |
| Listing management page | Bundle is a broader upgrade than one-off ad fixes |
| Paid-feature purchase flow | Bundle is a structured alternative to single visibility purchases |

Apply by tier:

| Tier | When | Frequency |
|---|---|---|
| 1 | All high-intent surfaces | Full prompt set, capped over first 2 weeks |
| 2 | Selected high-intent surfaces | Limited prompts, lower frequency |
| 3 | Lowest-friction surfaces only | One touch |


## 4. KPIs

### 4.1 Primary KPI

| KPI | Definition | Why it matters |
|---|---|---|
| **Bundle adoption rate** | Share of exposed sellers who adopt a bundle, overall and by tier | Measures whether the campaign drives bundle adoption. |

Use a holdout or A/B test where possible to estimate incremental adopters.

### 4.2 Secondary KPIs

| KPI                   | Definition | Why it matters |
|-----------------------|---|---|
| Click-through rate    | Share of exposed sellers who click through | Diagnoses channel and placement relevance. |
| Trial start rate      | Share of exposed sellers who start a bundle trial | Shows whether interest becomes action. |
| Bundle revenue        | Bundle revenue from campaign adopters | Links adoption to business value. |

### 4.3 Guardrail KPIs

| KPI | Definition | Why it matters |
|---|---|---|
| Early cancellation / downgrade rate | Share of adopters who cancel or downgrade early | Flags poor-fit or over-pushed sellers. |
| Seller opt-out or complaint rate | Share of exposed sellers who opt out or complain | Caps campaign pressure before trust suffers. |
