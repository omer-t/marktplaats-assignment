# Question 4: Marketing Campaign Proposal

## Assignment
> The introduction of SMB Bundles has provided value to both Marktplaats and our SMB sellers, as they are now more visible on our platform and are able to distinguish themselves from others.
>
> Without access to additional data, we would like you to think conceptually and answer the following:
>
> Imagine you are tasked with promoting the new SMB bundles more broadly. Propose one marketing campaign, including the target audience, the channel(s) you would use, and the key KPI(s) you would focus on to measure success.


## 1. Campaign Objective

- Adoption campaign for high-potential SMB sellers.

Potential future campaigns:
- Reactivation campaign for SMB sellers with declining performance.
- Upsell campaign from Basic to Plus.


## 2. Target Audience

- Active SMB sellers without an SMB bundle.
- First filter for sellers with a high SMB likelihood score (from question #1), so the campaign focuses on likely business sellers instead of consumers.
- Then prioritize within that audience using the Q2 readiness score (from question #2), because it captures recent activity, paid-product adoption, consistency, category breadth, growth, and fees.
- Start with sellers who score high on both SMB likelihood and bundle readiness; use Basic vs Plus positioning based on the Q2 bundle-fit recommendation.


## 3. Channels

### 3.1 Lifecycle CRM

Use email / CRM journeys triggered by seller behavior.

Examples of triggers:

| Trigger | Purpose |
|---|---|
| Seller lists repeatedly | Promote bundle when commercial intent is high |
| Seller uses paid features | Position bundle as a structured visibility product |
| Seller operates in competitive category | Promote differentiation benefit |
| Seller receives buyer interest but has no bundle | Promote visibility/professionalization as next step |

### 3.2 In-platform Prompts

Use prompts in high-intent seller surfaces:

| Surface | Why |
|---|---|
| Seller dashboard | Seller is already reviewing performance |
| Listing creation flow | Seller is investing effort in supply |
| Listing management page | Seller is optimizing existing ads |
| Paid-feature purchase flow | Seller is already considering visibility spend |


## 4. Key KPIs

### 4.1 Primary KPI

| KPI | Definition | Why it matters |
|---|---|---|
| **Incremental SMB Bundle adoption rate** | Difference in adoption between exposed sellers and a holdout/control group | Directly measures whether the campaign creates additional bundle adoption beyond what would have happened anyway. |

### 4.2 Secondary KPIs

| KPI                   | Definition | Why it matters |
|-----------------------|---|---|
| Click-through rate    | Share of exposed sellers who click through to bundle details or pricing | Shows whether the message and timing create genuine interest among the intended sellers. |
| Trial start rate      | Share of exposed sellers who begin the bundle sign-up flow | Identifies whether interested sellers are willing to take a concrete next step toward adoption. |
| Bundle revenue uplift | Incremental bundle revenue from exposed sellers vs. holdout/control group | Confirms that adoption translates into commercial impact, not only free or low-value sign-ups. |

### 4.3 Guardrail KPIs

| KPI | Definition | Why it matters |
|---|---|---|
| Early cancellation / downgrade rate | Share of new bundle adopters who cancel or downgrade shortly after adoption | Protects against pushing sellers into bundles that do not fit their needs or expectations. |
| Seller opt-out or complaint rate | Share of exposed sellers who unsubscribe, dismiss repeatedly, or contact support negatively | Ensures the campaign does not create excessive seller friction or damage trust. |
